Metoda optymalizacji przy pomocy biblioteki pulp pozwaliła również na analizę prostego obwodu elektrycznego i wyznaczenie maksymalnych wartości prądów dla poszczególnych gałęzi układu.

image.png

Dodatkowo został podany warunek:
Ii - deltai <= Ii <= Ii + deltai (gdzie "i" = 1...5 czyli prąd odpowiadający każdemu z kolejnych rezystorów)

```
from pulp import *

prob = LpProblem("Maksymalny prąd",LpMaximize)

#zmienne odpowiadające wszystkim prądom w gałęziach wraz z określeniem wartości wynikających z podanego warunku
I1=LpVariable("I1",0,2)
I2=LpVariable("I2",0,3)
I3=LpVariable("I3",0,4)
I4=LpVariable("I4",0,2)
I5=LpVariable("I5",0,2)

prob += I4 + I5, "prąd wyjściowy"

#warunki obliczeniowe dla obwodu wynikające z praw kirchhoff'a
#suma prądów wpływających do węzła równa się sumie prądów wypływających z węzła
prob += I1 + I2 == I3, "węzeł 1"
prob += I4 + I5 == I3, "węzeł 2"
#całkowita suma spadków napięć w obwodzie jest równa 0
prob += I1 * 8.0 - I2 * 6 == 0, "oczko 1"
prob += I4 * 10.0 - I5 * 8 == 0, "oczko 2"

prob.writeLP("Prady.lp")
prob.solve() # inicjalizacja funkcji optymalizacyjnej
print("Status:", LpStatus[prob.status]) #wyświetlenie statusu rozwiązania problemu
for v in prob.variables():
    print(v.name, "=", v.varValue, "A") #wypisanie nazw i wartości prądów
print()
print("Imax:",value(prob.objective),"A\n")  # wartość maksymalna prądu w obwodzie
```

Otrzymane wyniki wydają się być realne, co onzacza, że optymalizacja przebiegła pomyślnie. Dodając kolejne elementy układu i modyfikując ich parametry będziemy w stanie analizać bardziej skomplikowane układy.

**Zadanie 2**

Wyznaczenie minimalnej mocy rozporszonej wydzielonej przez układ z pierwszej częsci zadania. W tym zadaniu należało zadbać o to, aby suma mocy wydzielonej na każdym rezystorze była jak najmniejszą, co przełoży się na niski pobór prądu przez układ elektryczny.

image.png

```
from pulp import *

prob = LpProblem("Minimalizacja mocy utraconej",LpMinimize)

I1=LpVariable("I1",3,5)
I2=LpVariable("I2",1,3)
I3=LpVariable("I3",1,3)
I4=LpVariable("I4",1,3)
I5=LpVariable("I5",3,5)

U = (6.0, 10.0, 4.0, 7.0, 3.0) # wartość napięć na poszczególnych rezystorach

prob += (I1 * U[0]) + (I2 * U[1]) + (I3 * U[2]) + (I4 * U[3]) + (I5 * U[4]), "Moc całkowita"

#ograniczenia wynikające z praw Kirchhoff'a
prob += I1 - I3 == I4, "oczko górne"
prob += I2 + I3 == I5, "oczko dolne"

prob.writeLP("Moce.lp")
prob.solve()
print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue, "mA")
print()
print("Straty:",value(prob.objective),"mW")  
print()


for v, x in zip(U,prob.variables()):
  print("R", U.index(v)+1, "=", (v/x.varValue), "kOhm")
#wydruk wartości poszczególnych rezystorów
#warto pamiętać tutaj o odpowiedni dobór jednostki, w przeciwnym wypadku 
#możemy pomylić rząd wielkości elektrycznych
```

Uzyskana wartość całkowitej mocy rozproszonej wynosi 52mW.
